## Pearson method
scipy.stats

```python
# Pearson Method

import scipy.stats as st

# Correlation & P-value of Engine Size & Price
correlation, p_value = st.pearsonr(df['price'], df['engine-size'])
print(correlation, p_value)
```

```python
# Correlation & P-value of Stroke & Price
correlation, p_value = st.pearsonr(df['price'], df['stroke'])
print(correlation, p_value)
```

## Group By

```python
df1 = df[['body-style', 'drive-wheels', 'price']]
df1.head()
```

```python
df1_group = df1.groupby(['body-style', 'drive-wheels'], as_index=False).mean()
df1_group
```

```python
df1_group = df1.groupby(['body-style', 'drive-wheels']).mean()
df1_group
```

```python
df1_pivot = df1_group.pivot(index='body-style', columns='drive-wheels')
df1_pivot
```

```python
df1_pivot = df1_group.pivot(index='drive-wheels', columns='body-style')
df1_pivot
```

## Heatmap

```python
#heatmap

fig, ax = plt.subplots()
im = ax.pcolor(df1_pivot, cmap='RdBu')

#label names
row_labels = df1_pivot.columns.levels[1]
col_labels = df1_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(df1_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(df1_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

```python
df2 = df[['num-of-doors', 'drive-wheels', 'price']]
df2_group = df2.groupby(['num-of-doors', 'drive-wheels'], as_index=False).mean()
df2_pivot = df2_group.pivot(index='drive-wheels', columns='num-of-doors')

df2_pivot.head()
```

```python
#heatmap

fig, ax = plt.subplots()
im = ax.pcolor(df2_pivot, cmap='RdBu')

#label names
row_labels = df2_pivot.columns.levels[1]
col_labels = df2_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(df2_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(df2_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
