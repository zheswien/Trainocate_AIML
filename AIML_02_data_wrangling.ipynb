# Data Normalization (Do not mix the approaches in 1 dataset!)

```python
mydata['width'].head(20)
```

```python
mydata[['width', 'length', 'height']].head(10)
```

## Simple Feature Scaling

```python
mydata['width'] = mydata['width']/mydata['width'].max()
mydata[['width', 'length', 'height']].head(10)
```

## min-max

```python
mydata['length'] = (mydata['length'] - mydata['length'].min()) / (mydata['length'].max() - mydata['length'].min())
mydata[['width', 'length', 'height']].head(10)
```

## z-score

```python
mydata['height'] = (mydata['height'] - mydata['height'].mean()) / mydata['height'].std()
mydata[['width', 'length', 'height']].head(10)
```

# Data Binning

```python
df['price'] =  df['price'].astype('float')
df['price'].describe()
```

```python
bins_price = np.linspace(df['price'].min(), df['price'].max(), num=4) # 3 categories required 4 bins
bins_price
# Low - 5118 ~ 18545.33333333
# Average - 18545.33333333 ~ 31972.66666667
# High - 31972.66666667 ~ 45400
```

```python
label_names = ['Low price', 'Average price', 'High price']
df['price_Label'] = pd.cut(df['price'], bins_price, labels=label_names, include_lowest=True)
df[['price', 'price_Label']]
```
